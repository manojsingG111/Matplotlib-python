```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
xx,yy = np.meshgrid(x,y)
```

```python
z = xx**2 + yy**2
```

```python
z.shape
```

```python
fig = plt.figure(figsize=(15,10))
ax = plt.subplot(projection="3d")
p = ax.plot_surface(xx,yy,z,cmap="viridis")
fig.colorbar(p)
```

```python
balls=pd.read_csv("deliveries.csv")
```

```python
balls.sample(2)
```

```python
total_six_per_over=balls[(balls["ball"].isin([1,2,3,4,5,6])) & (balls["batsman_runs"]==6)]
```

```python
df=total_six_per_over.pivot_table(index="over",columns="ball",values="batsman_runs",aggfunc="count")
```

```python
balls["over"].unique()
```

```python
plt.figure(figsize=(25,10))
plt.imshow(df)
plt.yticks(balls["over"].unique())
plt.xticks(np.arange(0,6), list(range(1,7)))
plt.colorbar()
```
